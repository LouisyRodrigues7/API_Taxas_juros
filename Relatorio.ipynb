## üìò Descri√ß√£o dos Dados

Os dados referem-se √†s **taxas de juros m√©dias** cobradas mensal e anualmente pelas institui√ß√µes financeiras, por segmento de cliente e modalidade de cr√©dito. Os dados s√£o disponibilizados por meio da [API de Taxas de Juros do Banco Central](https://olinda.bcb.gov.br/olinda/servico/taxaJuros/versao/v2/odata/ConsultaUnificada).

---

## üîé Par√¢metros da API

| Par√¢metro      | Tipo    | Descri√ß√£o                                                                 |
|----------------|---------|---------------------------------------------------------------------------|
| `$format`      | texto   | Define o formato da resposta (ex: `json`, `xml`)                          |
| `$select`      | texto   | Define os campos a serem retornados (ex: `InstituicaoFinanceira`)         |
| `$filter`      | texto   | Filtros para refinar a consulta (ex: `Segmento eq 'Pessoa F√≠sica'`)       |
| `$orderby`     | texto   | Ordena√ß√£o dos dados (ex: `TaxaJurosAoAno desc`)                           |
| `$skip`        | inteiro | Define quantos registros ser√£o ignorados no in√≠cio (pagina√ß√£o)            |
| `$top`         | inteiro | Limita o n√∫mero de registros retornados                                   |

---

## üìà Campos Retornados

| Campo                  | Tipo      | Descri√ß√£o                                                              |
|------------------------|-----------|------------------------------------------------------------------------|
| `InicioPeriodo`        | texto     | Data de in√≠cio do per√≠odo (YYYY-MM-DD)                                |
| `FimPeriodo`           | texto     | Data de fim do per√≠odo (YYYY-MM-DD)                                   |
| `Segmento`             | texto     | Tipo de cliente: `Pessoa F√≠sica` ou `Pessoa Jur√≠dica`                 |
| `Modalidade`           | texto     | Tipo de cr√©dito (ex: Cart√£o, Cheque Especial, etc.)                   |
| `Posicao`              | inteiro   | Posi√ß√£o da institui√ß√£o no ranking de taxas                            |
| `InstituicaoFinanceira`| texto     | Nome da institui√ß√£o financeira                                         |
| `TaxaJurosAoMes`       | decimal   | Taxa m√©dia mensal (%)                                                  |
| `TaxaJurosAoAno`       | decimal   | Taxa m√©dia anual (%)                                                   |
| `cnpj8`                | texto     | Primeiros 8 d√≠gitos do CNPJ da institui√ß√£o                            |

````python
# Mostrar as primeiras linhas do DataFrame
df.head()
````

````python
# C√©lula 4: Converter colunas de datas para datetime
for col in ['InicioPeriodo', 'FimPeriodo']:
    df[col] = pd.to_datetime(df[col], errors='coerce')
````

````python
# Informa√ß√µes gerais do DataFrame e tipos de dados
df.info()
````

````python
# Estat√≠sticas descritivas b√°sicas das colunas num√©ricas
df[['TaxaJurosAoMes', 'TaxaJurosAoAno']].describe()
````

````python
#C√°lculo de m√©dia, vari√¢ncia e desvio padr√£o das taxas de juros
mean_mes = df['TaxaJurosAoMes'].mean()
var_mes = df['TaxaJurosAoMes'].var()
std_mes = df['TaxaJurosAoMes'].std()

mean_ano = df['TaxaJurosAoAno'].mean()
var_ano = df['TaxaJurosAoAno'].var()
std_ano = df['TaxaJurosAoAno'].std()

print(f"Taxa de Juros ao M√™s: M√©dia={mean_mes:.2f}%, Vari√¢ncia={var_mes:.4f}, Desvio Padr√£o={std_mes:.2f}")
print(f"Taxa de Juros ao Ano: M√©dia={mean_ano:.2f}%, Vari√¢ncia={var_ano:.4f}, Desvio Padr√£o={std_ano:.2f}")
````

````python
# Contagem das modalidades mais frequentes
df['Modalidade'].value_counts().head(10)
````

````python
# C√©lula 9: Filtrar dados s√≥ para 'Pessoa Jur√≠dica' ( Quantidade de linhas que cont√©m somente os dados de segnmento Pessoa jur√≠dica)
df_pj = df[df['Segmento'] == 'Pessoa Jur√≠dica']
print(f"N√∫mero de registros para Pessoa Jur√≠dica: {len(df_pj)}")
````

````python
#Estat√≠sticas descritivas para Pessoa Jur√≠dica (Taxas de Juros)
df_pj[['TaxaJurosAoMes', 'TaxaJurosAoAno']].describe()
````

````python
# vari√¢ncia e desvio padr√£o para Pessoa Jur√≠dica
mean_mes_pj = df_pj['TaxaJurosAoMes'].mean()
var_mes_pj = df_pj['TaxaJurosAoMes'].var()
std_mes_pj = df_pj['TaxaJurosAoMes'].std()

mean_ano_pj = df_pj['TaxaJurosAoAno'].mean()
var_ano_pj = df_pj['TaxaJurosAoAno'].var()
std_ano_pj = df_pj['TaxaJurosAoAno'].std()

print(f"[Pessoa Jur√≠dica] Taxa de Juros ao M√™s: M√©dia={mean_mes_pj:.2f}%, Vari√¢ncia={var_mes_pj:.4f}, Desvio Padr√£o={std_mes_pj:.2f}")
print(f"[Pessoa Jur√≠dica] Taxa de Juros ao Ano: M√©dia={mean_ano_pj:.2f}%, Vari√¢ncia={var_ano_pj:.4f}, Desvio Padr√£o={std_ano_pj:.2f}")
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
````

## Quais modalidades de cr√©dito t√™m as maiores m√©dias de taxa de juros anuais entre pessoas jur√≠dicas.

````python
pj_modalidade_taxa = df_pj.groupby('Modalidade')['TaxaJurosAoAno'].mean().sort_values(ascending=False)
pj_modalidade_taxa.head(10)
````

## Estudo por gr√°ficos 

para complementar o processo de an√°lise de dados vamos gerar alguns gr√°ficos, para isso usaremos a biblioteca matplotlib.



## Histograma das taxas de juros ao m√™s (todas as modalidades)

````python
plt.hist(df['TaxaJurosAoMes'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histograma da Taxa de Juros ao M√™s (todas modalidades)')
plt.xlabel('Taxa de Juros ao M√™s (%)')
plt.ylabel('Frequ√™ncia')
plt.show()
````

## Boxplot das taxas de juros ao ano agrupadas por modalidade (top 10 modalidades)

````python
top_modalidades = df['Modalidade'].value_counts().head(10).index
df_top_modalidades = df[df['Modalidade'].isin(top_modalidades)]

plt.figure(figsize=(14,7))
sns.boxplot(x='TaxaJurosAoAno', y='Modalidade', data=df_top_modalidades, palette='Set3')
plt.title('Boxplot da Taxa de Juros ao Ano para as 10 Modalidades mais comuns')
plt.xlabel('Taxa de Juros ao Ano (%)')
plt.ylabel('Modalidade')
plt.show()
````

## S√©rie temporal da taxa m√©dia mensal ao longo do tempo para Pessoa Jur√≠dica

````python
df_time_pj = df_pj.groupby('InicioPeriodo')['TaxaJurosAoMes'].mean()

plt.figure(figsize=(14,6))
df_time_pj.plot()
plt.title('Evolu√ß√£o da Taxa M√©dia Mensal ao Longo do Tempo (Pessoa Jur√≠dica)')
plt.xlabel('Data de In√≠cio do Per√≠odo')
plt.ylabel('Taxa M√©dia Mensal (%)')
plt.grid(True)
plt.show()
````

## Gr√°fico da m√©dia da taxa anual por modalidade para Pessoa Jur√≠dica (top 10)

````python
plt.figure(figsize=(12,6))
pj_modalidade_taxa.head(10).plot(kind='bar', color='coral')
plt.title('M√©dia da Taxa Anual por Modalidade (Pessoa Jur√≠dica) - Top 10')
plt.ylabel('Taxa M√©dia Anual (%)')
plt.xlabel('Modalidade')
plt.xticks(rotation=45, ha='right')
plt.show()
````
